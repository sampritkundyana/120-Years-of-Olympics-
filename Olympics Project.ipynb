```python
# Loading the datasets we have (athlete_events and noc_regions)
athletes = pd.read_csv('C:/Users/user/Desktop/Project/athlete_events.csv')
athletes.head()
```

```python
# Loading another datasets called regions
regions = pd.read_csv('C:/Users/user/Desktop/Project/noc_regions.csv')
regions.head()
```

#### We have two dataset Athletes and Region

```python
# Now we will combine this two dataset for further Analysis
# Regions data set has region and notes, We will combine this dataset to athletes dataframe.
```

```python
# Using inner joins we will combine regions datasets with athletes dataframe.
ath = athletes.merge(regions, how='inner', on='NOC')
ath.head()
```

```python
# Region and notes has been merged in ath
```

```python
# Checking Top 10 countries who have participated in Olympics Till 2016
Top_10_Countries=ath.Team.value_counts().sort_values(ascending=False).head(10)
Top_10_Countries
```

- United States has highest athletes participated in Olympics followed France and Great Britain

```python
# Checking top 10 Countries who  have participated in Olympics
plt.figure(figsize=(15,9))
plt.title('Countries Participated in Olympics', fontdict={'fontsize':15})
plt.ylabel('Team', fontdict={'fontsize':15})
plt.xlabel('Countries', fontdict={'fontsize':15})
sns.barplot(x=Top_10_Countries.index, y=Top_10_Countries, palette='inferno')
```

```python
# Checking Number of Males and Females participated in Olympics
Gender=ath['Sex'].value_counts()
Gender
```

```python
# Now we will Plot pie chart for gender 
plt.figure(figsize=(10,7))                                                             # Figsize
plt.title('Number of Males and Females Participation in olympics in Percentage')       # Title
plt.pie(Gender, labels=Gender.index, autopct='%1.1f%%', startangle=90)  # autopct will help us to display the percent values using python string formatting
                                                                           
```

- Here 72.5% of Mens have Participated in Olympics
- And 27.5% of Womens have Participated in Olympics 

```python
# Checking number of medal counts
medal_count=ath.Medal.value_counts()
medal_count
```

```python
# Plotting Medal counts in pie chart
plt.figure(figsize=(10,7))
plt.title("Medal Counts in Olympics")
sns.countplot(ath.Medal.value_counts())
plt.show()
```

```python
# Plotting histplot for age Category
plt.figure(figsize=(15,9))
plt.title('Athletes Age Distribution', fontdict={'fontsize':15})
plt.xlabel('Age', fontdict={'fontsize':15})
plt.ylabel('Number', fontdict={'fontsize':15})
sns.histplot(data=ath, x='Age', bins=np.arange(10,80), edgecolor='white')
```

- Athletes Age Between 20-30 Have participated in large Number

```python
# Types of Sports that belongs to Winter Olympics
winter_sport=ath.query("Season == 'Winter'")['Sport'].unique()         # we used query to filter the dataframe
winter_sport
```

```python
# Lets Count number of Athletes participated in winter Olympics
plt.figure(figsize=(15,9))
ath.query("Season == 'Winter'")['Sport'].value_counts().plot(kind='bar')
plt.show()
```

- Cross Country skiing has the highest number of Participants

```python
# Lets check number of Participants of Male and female for winter Sports
plt.figure(figsize=(15,9))
sns.countplot(data=ath, y=ath.query("Season == 'Winter'")['Sport'], hue='Sex', palette='gist_heat')
```

```python
# Checking number of Sports that is included in Summer Olympic
summer_sport=ath.query("Season == 'Summer'")['Sport'].unique()
summer_sport
```

```python
# Checking the Summer Olympics using barchart
plt.figure(figsize=(15,9))
ath.query("Season == 'Summer'")['Sport'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Checking For Cities where Olympics games held
plt.figure(figsize=(12,6))
plt.title('Cities where Olympics Held')
plt.xlabel('City')
plt.ylabel('Year')
plt.xticks(rotation=90)
sns.scatterplot(data=ath, x='City', y='Year')
```

```python
# Checking Countries who bagged medals for the year 2016 in Rio olympics
plt.figure(figsize=(10,9))
sns.barplot(x=Gold_team.value_counts().head(10), y=Gold_team.value_counts().head(10))
plt.xlabel('Countries Bagged Medals for year 2016')
plt.ylabel('Medals Count')
plt.show()
```
